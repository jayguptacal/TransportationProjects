### Conventional Crude Oil for G7<a id='3.4.1.1_Conventional_Crude_Oil_for_G7'></a>

```python
# Analyzing conventional crude oil for G7 countries
crudeoil_g7 = group7[group7.category.isin(['conventional_crude_oil'])].sort_values(['year', 'country_or_area'])
```

```python
crudeoil_g7.head()
```

```python
# drilling down further into the commodity transaction now to calculate conventional crude production and imports
crudeprod_g7 = crudeoil_g7[crudeoil_g7.commodity_transaction.isin(['Conventional crude oil - production'])].sort_values(['year', 'country_or_area'])
crudeimport_g7 = crudeoil_g7[crudeoil_g7.commodity_transaction.isin(['Conventional crude oil - imports'])].sort_values(['year', 'country_or_area'])
```

```python
# checking for the crude oil production for the G7 countries
crudeprod_g7.head()
```

```python
# checking for the crude oil imports for the G7 countries
crudeimport_g7.head()
```

```python
# combining crude oil production and imports
crudereq_g7 = pd.concat([crudeprod_g7, crudeimport_g7]).sort_values(['country_or_area'])
```

```python
crudereq_g7.head()
```

```python
crudereq_can = crudereq_g7[crudereq_g7.country_or_area.isin(['Canada'])].sort_values('year')
```

```python
crudereq_can.head()
```

```python
crudepivot_can = crudereq_can.pivot_table(index='year', columns='commodity_transaction')
```

```python
crudepivot_can.head()
```

```python
# Plotting crude oil production and import for Canada over the time format
crudepivot_can.plot(kind='line')
plt.title("Conventional Crude Oil Production and Import for Canada")
plt.xlabel("Year")
plt.ylabel("Total Quantity (Metric tons, thousand)")
plt.legend(loc='best')
plt.show()
```

The above chart confirms that Canada has been increasing Crude Oil production significantly over the time and that's why imports has been almost constant.

```python
# Plotting Conventional Crude Oil production and import for USA
crudereq_us = crudereq_g7[crudereq_g7.country_or_area.isin(['United States'])].sort_values('year')
crudepivot_us = crudereq_us.pivot_table(index='year', columns='commodity_transaction')

# Plotting crude oil production and import for USA over the time format

crudepivot_us.plot(kind='line')
plt.title("Conventional Crude Oil Production and Import for USA")
plt.xlabel("Year")
plt.ylabel("Total Quantity (Metric tons, thousand)")
plt.legend(loc='best')
plt.show()
```

### Nuclear Electricity for G7<a id='3.4.1.2_Nuclear_Electricity_for_G7'></a>

```python
# making the column names easier by renaming the column labels
energy.columns = ['country', 'commodity', 'year', 'unit', 'quantity', 'category']
energy.head()
```

```python
def select_country(countryname):
    data = energy[energy.country.isin(countryname)].sort_values('year').reset_index()
    data.drop('index', axis=1, inplace=True)
    return data
```

```python
def select_category(country_data, categoryname):
    countrycom_data = country_data[country_data.category.isin(categoryname)].sort_values('year').reset_index()
    countrycom_data.drop('index', axis=1, inplace=True)
    return countrycom_data
```

```python
# Nuclear Electricity for Canada for each year to plot it together with other G7 Countries

can_data = select_country(['Canada'])
nuclear_can = select_category(can_data, ['nuclear_electricity'])
x1 = nuclear_can.year
y1 = nuclear_can.quantity

# Nuclear Electricity for USA for each year to plot it together with other G7 Countries
us_data = select_country(['United States'])
nuclear_us = select_category(us_data, ['nuclear_electricity'])
x2 = nuclear_us.year
y2 = nuclear_us.quantity

# Nuclear Electricity for UK for each year to plot it together with other G7 Countries
uk_data = select_country(['United Kingdom'])
nuclear_uk = select_category(uk_data, ['nuclear_electricity'])
x3 = nuclear_uk.year
y3 = nuclear_uk.quantity

# Nuclear Electricity for Germany for each year to plot it together with other G7 Countries
germany_data = select_country(['Germany'])
nuclear_germany = select_category(germany_data, ['nuclear_electricity'])
x4 = nuclear_germany.year
y4 = nuclear_germany.quantity

# Nuclear Electricity for France for each year to plot it together with other G7 Countries
france_data = select_country(['France'])
nuclear_france = select_category(france_data, ['nuclear_electricity'])
x5 = nuclear_france.year
y5 = nuclear_france.quantity

# Nuclear Electricity for Italy for each year to plot it together with other G7 Countries
italy_data = select_country(['Italy'])
nuclear_italy = select_category(italy_data, ['nuclear_electricity'])
x6 = nuclear_italy.year
y6 = nuclear_italy.quantity

# Nuclear Electricity for Japan for each year to plot it together with other G7 Countries
japan_data = select_country(['Japan'])
nuclear_japan = select_category(japan_data, ['nuclear_electricity'])
x7 = nuclear_japan.year
y7 = nuclear_japan.quantity
```

```python
# Plotting the Nuclear Electricity line chart for the Group 7 Countries together

plt.figure(figsize=(10, 8))
plt.plot(x1, y1, label='Canada')
plt.plot(x2, y2, label='United States')
plt.plot(x3, y3, label='United Kingdom')
plt.plot(x4, y4, label='Germany')
plt.plot(x5, y5, label='France')
plt.plot(x6, y6, label='Italy')
plt.plot(x7, y7, label='Japan')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Nuclear Electricity (Kilowatt-hours, million)", fontsize=12)
plt.title("Nuclear Electricity for the G7 Countries")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(loc='best')
plt.show()
```

```python
# Plotting the Nuclear Electricity bar chart for the Group 7 Countries together

plt.figure(figsize=(12, 8))
plt.bar(x1, y1, label='Canada')
plt.bar(x2, y2, label='United States')
plt.bar(x3, y3, label='United Kingdom')
plt.bar(x4, y4, label='Germany')
plt.bar(x5, y5, label='France')
plt.bar(x6, y6, label='Italy')
plt.bar(x7, y7, label='Japan')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Nuclear Electricity (Kilowatt-hours, million)", fontsize=12)
plt.title("Nuclear Electricity for the G7 Countries")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(loc='best')
plt.show()
```
